## Advance operation with business examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  #Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  #Fresh Bites
    [3, 200000, 230000, 260000, 300000],  #Kukas King
    [4, 180000, 210000, 240000, 270000],  #Out of Town
    [5, 160000, 185000, 205000, 230000]   #Pizza Hub

])

print("===Zomato Sales Analysis===")
print("sales data shape: ",sales_data.shape)
print("Sales data for first 3 restaurants: \n",sales_data[:3])
```

```python
#total sales per year
print("Yearly sum but with res id also: ",np.sum(sales_data,axis=0))# yearly sum with res_id

total_sum_yearly = np.sum(sales_data[:,1:],axis=0)
print("Yearly sum: ",total_sum_yearly)

#min sales per restaurant
print("min sales of restaurants: ",np.min(sales_data[:,1:],axis=1))

#max sales per year
print("max sales per year(not of restaurant:\n)",np.max(sales_data[:,1:],axis=0))

#aveg sales per retaurant
mean = np.mean(sales_data[:,1:],axis=1)
print("Average sales/restaurant:\n",mean)

```

```python
#cumalative sum: adding xi from i = 1 to n
cumsum = np.cumsum(sales_data[:,1:],axis=1)
print("Cumalative sum of res:\n",cumsum)

plt.figure(figsize=(5,3))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

```

### vectors operations

```python
vec1 = np.array([1,2,3])
vec2 = np.array([4,5,7])

print("addition: ",vec1+vec2)#OR np.add(vec1,vec2))

print("Multiplication: ",np.multiply(vec2,vec1)) # or vec1 * vec2

print("Scalar(Dot) product: ",np.dot(vec1,vec2))


#  cos@ = a.b / |a||b|
angle = np.arccos(np.dot(vec1,vec2) / (np.linalg.norm(vec1) * np.linalg.norm(vec2)))
print("angle: ",angle)

```

### make an operation vectorised
 means it can be compatible to numpy array to perform operation on all elements instead of one

```python
res_name = np.array(["Biryani","Chinese", "Momos"])
vectorized_uppper = np.vectorize(str.upper)# making upper method compatible with whole array,instead of single element
print("Vectorized upper: ", vectorized_uppper(res_name))
```

### broadcast
  scalars dont need to vectorize, numpy is smart enough to handle this
  

```python
monthly_avg = np.array(sales_data[:,1:]) / 12 # here /12 is scalar operation, so we dont need to vectorize it
print("avg: ",monthly_avg)
```
