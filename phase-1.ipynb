## Numpy array and basics
learn about numpy

```python
import numpy as np
```

## Creating array from list


```python
arr_1d = np.array([1,2,3,4,5])
print("1D array: ",arr_1d)

arr_2d = np.array([[1,2,3],[4,5,6]])
print("2D array:",arr_2d)
```

## List vs numpy array

```python
py_list = [1,2,3]
print("list multiplication by 2:",py_list * 2)

py_array = np.array([1,2,3])#element wise multiplication
print("array multiplication by 2:",py_array * 2)

import time
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation Time: ", time.time()-start)

start = time.time()
py_array = np.arange(100000) * 2
print(py_array)
print("\n array operation Time: ", time.time()-start)
```

### Creating array from scratch


```python
zero = np.zeros((2,3))
print("zeroes array: ")
print(zero)
print()

one = np.ones((2,4))
print("ones array: ")
print(one)
print()

full = np.full((2,2),4)
print("fulls array:")
print(full)
print()

random = np.random.random((2,4)) * 10
random = random - random%1 #skipping decimal part
print(random)
```

### Vector, Matrix and Tensor

```python
vector = np.array([1,2,3])
print("vector:")
print(vector)
print()

matrix = np.array([[2,3],
                   [3,4]])
print("Matrix:\n",matrix)

print()

tensor = np.array([[1,2],[2,3],
                   [4,5],[5,6]])
print("tensors:\n",tensor)
```

### Array Properties

```python
arr = np.array([[1,2,3],
                [4,5,6]])
print("Shape: ",arr.shape)
print("Dimension: ",arr.ndim)
print("Size: ",arr.size)
print("DType: ",arr.dtype)


```

### Array Reshaping

```python
original = np.arange(12)
print("original array: \n",original)

reshaped = original.reshape((3,4))
print("reshaped array: \n",reshaped)

flattened = reshaped.flatten()
print("flattend : \n", flattened)

raveled = reshaped.ravel()#returns view instead of copy
print("raveled: \n",raveled)

#transpose
transpose = reshaped.T
print("transposed: \n",transpose)
```
